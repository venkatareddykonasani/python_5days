# Merging the datasets

```
#Import the AB_NYC_2019_Reviews.csv data. Attach ratings data to the input data.
review_data=pd.read_csv("https://raw.githubusercontent.com/venkatareddykonasani/python_5days/master/Day2_Files/Datasets/AB_NYC_2019_Reviews.csv")
review_data.shape
```

```
review_data.info()
```

```
input_and_review_l=pd.merge(input_data,review_data, on=["id"], how="left")
```

```
print("input_data shape", input_data.shape)
print("review_data shape", review_data.shape)
print("input_and_review_l shape", input_and_review_l.shape)
```

```
input_and_review_inner=pd.merge(input_data,review_data, on=["id"], how="inner")
print("input_data shape", input_data.shape)
print("review_data shape", review_data.shape)
print("input_and_review_inner shape", input_and_review_inner.shape)
```

```
input_and_review_right=pd.merge(input_data,review_data, on=["id"], how="right")
print("input_data shape", input_data.shape)
print("review_data shape", review_data.shape)
print("input_and_review_right shape", input_and_review_right.shape)
```

```
input_and_review_outer=pd.merge(input_data,review_data, on=["id"], how="outer")
print("input_data shape", input_data.shape)
print("review_data shape", review_data.shape)
print("input_and_review_outer shape", input_and_review_outer.shape)
```

# Exporting the data
